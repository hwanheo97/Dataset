## 오늘의 핵심강의 : 'Yes24 국내 도서 종합 베스트셀러' 크롤링 코드 이해 및 활용

아래 두개 셀을 여러분 실습을 돕기위해서 추가로 작성된 부분이니 사전에 충분히 이해하고 실습 영상을 보시기 바랍니다.

기존 교보문고 주간베스트 코드가 변경되어 여러분이 이해가능한 사이트를 참조하여 유사하게 Yes24 국내도서를 참조하여 작성하였습니다.
설명에 참조되는 교보문고 사이트는 아래의 링크를 들어가서 참고하세요. 

* 교보문고 링크 : https://product.kyobobook.co.kr/bestseller/total?period=002#?page=1&per=20&period=002&ymw=&bsslBksClstCode=A

사이트로 이동하여 소스코드를 보는 방법은 위에 링크를 누르고 우측마우스 클릭 - 페이지소스보기를 눌러서 보는 방법도 있고


기능기 F12를 눌러서 보면 오른쪽으로 소스 코드가 나오니 그것을 보시면 됩니다. 

* Yes24 링크 : http://www.yes24.com/24/Category/BestSeller?CategoryNumber=001&sumgb=06


HTML 소스코드를 F12로 열고 해당 소스코드로 찾아들어가는 방법은 소스코드의 좌상단 마우스 화살표 부분을 마우스로 클릭하고 좌측 본래의 문서로 들여가면 음영으로 반전되면서 해당 부분에 대한 tag명.class명이 나타납니다. 그부분의 소스코드를 보고자하면 우측 마우스를 클릭하면 해당 소스코드 부분이 고정 음영이 반전되어 표시되니 그 소스코드를 열어서 자세히 보시면 됩니다. 이러한 기능을 잘 활용하시면 실습이 용이하니 영상에서 실습부분을 실행하시기 전에 이기능을 이해하시고 시작하시길 바랍니다. 

```python
# 사이트 접근하여 데이터 가져오기
import requests
html = requests.get('http://www.yes24.com/24/Category/BestSeller?CategoryNumber=001&sumgb=06')
print(html.encoding)
print(html.headers)
print(html.text[:200])
```

```python
# 정보를 크롤링하여 정리하기 : 약 1분 정도 대기해야 실행완료 됩니다.
from urllib.request import urlopen
from bs4 import BeautifulSoup as bs
import pandas as pd

html = urlopen('http://www.yes24.com/24/Category/BestSeller?CategoryNumber=001&sumgb=06')
soup = bs(html, "html.parser")
table = soup.find('table', class_='list')

book_urls =[]
for cont in table.find_all('div', {'class' : 'goodsImgW'}):
    link = 'http://www.yes24.com/'+cont.find('a').get('href')
    book_urls.append(link)

dic = []
for book_url in book_urls:
    html = urlopen(book_url)
    soup = bs(html, "html.parser")
    title = soup.find('h2', {'class':'gd_name'}).text
    author = soup.find('span', {'class': 'gd_auth'}).find('a').text
    pub = soup.find('span', {'class': 'gd_pub'}).text
    price = soup.find('span', {'class':'nor_price'}).text
    dic.append({'제목':title,'저자' : author,'출판사' : pub, '가격':price, 'Link' : book_url} )

pd.DataFrame(dic)
```

## 가.select와 find 사용법 차이

* select는 CSS 선택자 이용
* find는 태그와 속성, 속성값 이용
* find = select_one, find_all = select

```python
# find와 select_one 사용법 비교
from bs4 import BeautifulSoup as bs
tag = '''
<div class = 'section'>
<p class='start' id='s1'> 시작합니다! </p>
<p class='end' id='s2'> 마지막입니다! </p>
</div>
'''
soup = bs(tag) 

# find는 태그 이름, 속성으로 특정 
print(soup.find('p')) 
print(soup.find(class_='end')) 
print(soup.find(attrs = {'class':'start'})) 
print(soup.find('p', {'id':'s2'}))

# # select_one는 css로 특정
print(soup.select_one('p')) 
print(soup.select_one('.start')) 
print(soup.select_one('p.end')) 
print(soup.select_one('p#s1'))
print(soup.select_one('#s2'))
```

```python
# find_all과 select 사용법 비교
from bs4 import BeautifulSoup as bs

soup = bs(tag) 

print(soup.find('div').find('p').text)
print(soup.select_one('div > #s2').text)
for i in soup.find_all('p'):
  print(i.get('id'))
for i in soup.select('p') :
  print(i.get('id'))
```

# 5.사례 연구 : Yes24 국내 도서 종합 베스트셀러

## 가.HTML을 단계적으로 찾아들어가기

```python
# 필요한 라이브러리 선언, urlopen과 bs로 parsing
from urllib.request import urlopen
from bs4 import BeautifulSoup as bs
import pandas as pd

html = urlopen('http://www.yes24.com/24/Category/BestSeller?CategoryNumber=001&sumgb=06')
soup = bs(html, "html.parser")
table = soup.find('table', class_='list')
# print(table.prettify())
print(table.text[:200])
```

```python
# 베스트셀러의 제목을 찾아 출력 하기 
for i, cont in enumerate(table.find_all('div', {'class' : 'goodsImgW'})):
    if i >2 : break
    print(cont)
    # print(cont.find('img').get('alt'))
    print()
```

```python
#url 찾아내기, 출력된 링크를 열면 해당 책자에 대한 사이트로 연결
book_urls =[]
for i, cont in enumerate(table.find_all('div', {'class' : 'goodsImgW'})):
    if i >2 : break
    link = 'http://www.yes24.com/'+cont.find('a').get('href')
    print(link)
    book_urls.append(link)
book_urls
```

## 나.정리된 URL을 찾아들어가 정보를 가져오기

```python
# url별로 순회하면서 딕셔너리 데이터로 담아서 list에 담기
dic = []
for index, book_url in enumerate(book_urls):
    html = urlopen(book_url)
    soup = bs(html, "html.parser")
    title = soup.find('h2', {'class':'gd_name'}).text
    author = soup.find('span', {'class': 'gd_auth'}).find('a').text
    pub = soup.find('span', {'class': 'gd_pub'}).text
    price = soup.find('span', {'class':'nor_price'}).text
    print(title, author, price)
    dic.append({'제목':title,'저자' : author,'출판사' : pub, '가격':price, 'Link':book_url})
```

```python
# 데이터 확인
dic
```

```python
# 판다스 데이터 프레임으로 전환
pd.DataFrame(dic)
```
